```
enc = LabelEncoder()
enc.fit(output)
enc_y = enc.transform(output)
d_y = np_utils.to_categorical(enc_y)
```

model architecture for validation

```
def iris_model():
	# create model


	model = Sequential()
	model.add(Dense(8, input_dim=4, activation='relu'))
	model.add(Dense(3, activation='softmax'))
 

	# Compile model
	model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
	return model
```

```
estimator = KerasClassifier(build_fn=iris_model, epochs=200, batch_size=5, verbose=0)
```

kfold validation

```
kfold = KFold(n_splits=10, shuffle=True)
```

```
results = cross_val_score(estimator, input, d_y, cv=kfold)

print((results.mean()*100, results.std()*100))
```

```
def runAlgo(X_train,X_test,y_train,y_test):
    logi = Sequential()
    logi.add(Dense(8, activation="relu",input_shape=(4,)))
    logi.add(Dropout(0.3))
    logi.add(Dense(3,activation='softmax'))
    logi.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    logi.fit(X_train, y_train, epochs=1000)
    a = logi.evaluate(X_test, y_test)
    return a
```
